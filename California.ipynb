# Limpieza por cuartiles

```python
%matplotlib inline
df.boxplot('median_income')
```

```python
sP = df['median_income']
IQR = 1.5*(sP.quantile(.75)-sP.quantile(.25))
lim_sup = sP.quantile(.75)+IQR
lim_inf = sP.quantile(.25)-IQR
print('IQR: ', IQR, '\nQ75: ', sP.quantile(.75), '\nQ15: ', sP.quantile(.25), '\nlim_inf: ', lim_inf, '\nlim_sup: ', lim_sup)
```

```python
sP_clean = sP[(sP >= lim_sup) | (sP <= lim_inf)]
sP_clean
```

```python
df.drop(sP_clean.index)
```

```python
med_inc = df['median_income']
for k in range (len(med_inc)):
    if (med_inc[k] < 0.):
        print(med_inc[k], '\n')
```

# Limpieza por cuartiles

```python
from scipy import stats
sP = df['median_income']
z = np.abs(stats.zscore(sP))
print(z, '\n')
sP_clean = sP[(z>3) | (z<-3)]
df.drop(sP_clean.index)
```
